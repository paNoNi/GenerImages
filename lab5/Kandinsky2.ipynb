```python
from kandinsky2 import get_kandinsky2

model = get_kandinsky2('cuda', task_type='text2img')
```

```python
from PIL import Image


def get_images(text: str) -> Image:
    return model.generate_text2img(text, batch_size=1, h=512, w=512, num_steps=75, denoised_type='dynamic_threshold',
                                   dynamic_threshold_v=99.5, sampler='ddim_sampler', ddim_eta=0.05, guidance_scale=10)
```

```python
images = get_images('Стены и двери своих жилищ ведьмаки украшали шкурами забитых на охоте зверей – оленей, рысей, даже росомах. На дверях же комнатки Цири висела шкура гигантской крысы с отвратительным чешуйчатым хвостом.')

images[0]
```
